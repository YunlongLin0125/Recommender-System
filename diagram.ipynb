```python
import matplotlib.pyplot as plt

# Data from the table
learning_rates = [0.0005, 0.001, 0.005]
models = ['SASRec', 'SASRec-Window', 'All action', 'Dense all action']
recall_at_10 = [
    [0.4577, 0.4615, 0.4548 ],  # SASRec
    [0.4458, 0.4487, 0.4466 ],  # SASRec-Sampled
    [0.3412, 0.3459, 0.3340 ],  # All action
    [0.2304, 0.2298, 0.1980 ],  # Dense all action
]

# Plotting the data
plt.figure(figsize=[10, 6])
for model, recall in zip(models, recall_at_10):
    plt.plot(learning_rates, recall, marker='o', label=model)

plt.xlabel('Learning Rate')
plt.ylabel('Recall@10')
plt.title('Model Performance with Different Learning Rates on Recall@10')
plt.legend()
plt.grid(True)
plt.savefig('learning_rate.png')
plt.show()
```
